```python
print('驗證一組身分證字號是否有效')


idal=str(input('請輸入身分證字號第一碼英文字:'))
idnumber=str(input('請輸入身分證字號九碼數字(第一碼須為1或2):'))

newid= []

for p in idnumber:
    p2 = int(p)
    newid.append(p2)

s=0

for a in range(0,8):
    s=s+newid[a]*(8-a)

    
if idal=='A':
    s=s+1
elif idal=='B':
    s=s+10
elif idal=='C':
    s=s+19
elif idal=='D':
    s=s+28
elif idal=='E':
    s=s+37
elif idal=='F':
    s=s+46
elif idal=='G':
    s=s+55
elif idal=='H':
    s=s+64
elif idal=='I':
    s=s+39
elif idal=='J':
    s=s+73
elif idal=='K':
    s=s+82
elif idal=='L':
    s=s+2
elif idal=='M':
    s=s+11
elif idal=='N':
    s=s+20
elif idal=='O':
    s=s+48
elif idal=='P':
    s=s+29
elif idal=='Q':
    s=s+38
elif idal=='R':
    s=s+47
elif idal=='S':
    s=s+56
elif idal=='T':
    s=s+65
elif idal=='U':
    s=s+74
elif idal=='V':
    s=s+83
elif idal=='W':
    s=s+21
elif idal=='X':
    s=s+3
elif idal=='Y':
    s=s+12
elif idal=='Z':
    s=s+30
else:
    s=s+0


ninth=10-s%10

if ninth-newid[8]==0 :
   print("輸入的身分證號碼",idal+idnumber,"是有效的")
else :

   print("輸入的身分證號碼",idal+idnumber,"是無效的") 
```
