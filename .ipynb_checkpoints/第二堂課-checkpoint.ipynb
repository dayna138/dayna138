````python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
````

````python
π=np.pi
````

````python
θ=np.linspace(0,2*π,500)
````

````python
r=3
x=r*np.cos(θ)
y=r*np.sin(θ)
````

````python
plt.plot(x,y)
````

````python
ax=plt.gca()
ax.set_aspect('equal')

plt.plot(x,y)
````

````python
r=1-np.sin(θ)
x=r*np.cos(θ)
y=r*np.sin(θ)
````

````python
ax=plt.gca()
ax.set_aspect('equal')

plt.plot(x,y,'r')
````
